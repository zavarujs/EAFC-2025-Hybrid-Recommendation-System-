```python
from sklearn import preprocessing

x = df.values
scaler = preprocessing.MinMaxScaler()
x_scaled = scaler.fit_transform(x)
X_norm = pd.DataFrame(x_scaled)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 2) # 2D PCA untuk visual plot
reduced = pd.DataFrame(pca.fit_transform(X_norm))
```

```python
from sklearn.cluster import KMeans

# tentukan jumlah cluster
kmeans = KMeans(n_clusters=5) # 5 cluster cuk wakeh men
# ngopo kok dadak 5 cluster yo mergo iki kan data set pemain
# pasti due 5 posisi pasti : forward, midfield, defender, winger, goalkepeer. ngeten cok

# sesuai dengan data masukan
kmeans = kmeans.fit(reduced)

# mendapatkan cluster labels
labels = kmeans.predict(reduced)

# centroid values, nggo nemukke centroid
centroid = kmeans.cluster_centers_

# cluster values
clusters = kmeans.labels_.tolist() # range 0-4 cluster (1-5 cluster)
```

```python
reduced['cluster'] = clusters
reduced['name'] = names
reduced.columns = ['x', 'y', 'cluster', 'name'] # x dan y itu untuk reduksi PCAne jadi di kurangi dadi 2 dimensi
# yaitu x dan y ben visual neng 2 rung demensi
reduced.head()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics.pairwise import cosine_similarity
from scipy.stats import percentileofscore
import logging

class StrikerRecommender:
    """
    Recommender system for finding top strikers using hybrid scoring
    """
    def __init__(self, data):
        try:
            self.features = ['OVR', 'PAC', 'SHO', 'Acceleration', 'Positioning', 'Finishing', 'Shot Power', 'Composure', 'Heading Accuracy', 'Jumping']
            self.weights = {
                'OVR': 0.2, 'PAC': 0.15, 'SHO': 0.15,
                'Acceleration': 0.1, 'Positioning': 0.1,
                'Finishing': 0.1, 'Shot Power': 0.1,
                'Composure': 0.05, 'Heading Accuracy': 0.05,
                'Jumping': 0.05
            }
            self.df = data[['Name'] + self.features]
            self.scaler = MinMaxScaler()
            self.normalized_features = self.scaler.fit_transform(self.df[self.features])
            logging.info("Recommender initialized successfully")
        except Exception as e:
            logging.error(f"Initialization error: {e}")
            raise

    def calculate_scores(self):
        """Calculate content and collaborative scores"""
        try:
            content_scores = np.zeros(len(self.df))
            collaborative_scores = np.sum(cosine_similarity(self.normalized_features), axis=1)

            for i, feature in enumerate(self.features):
                normalized_feature = self.normalized_features[:, i]
                percentile_scores = np.array([percentileofscore(normalized_feature, val) for val in normalized_feature])
                content_scores += percentile_scores * self.weights[feature]

            return content_scores / len(self.features), collaborative_scores
        except Exception as e:
            logging.error(f"Score calculation error: {e}")
            raise

    def recommend(self, min_ovr=86, top_n=5):
        """Generate top striker recommendations"""
        try:
            content_scores, collaborative_scores = self.calculate_scores()
            hybrid_scores = (
                (content_scores - content_scores.min()) / (content_scores.max() - content_scores.min()) * 0.6 +
                (collaborative_scores - collaborative_scores.min()) / (collaborative_scores.max() - collaborative_scores.min()) * 0.4
            )

            self.df['Hybrid_Score'] = hybrid_scores
            recommended = self.df[self.df['OVR'] > min_ovr].sort_values('Hybrid_Score', ascending=False).head(top_n)
            recommended['Content_Score'] = content_scores[recommended.index]
            recommended['Collaborative_Score'] = collaborative_scores[recommended.index]

            logging.info(f"Generated top {top_n} recommendations")
            return recommended
        except Exception as e:
            logging.error(f"Recommendation generation error: {e}")
            raise

def main():
    logging.basicConfig(level=logging.INFO)
    try:
        df = pd.read_csv('/content/drive/MyDrive/SKRIPSI/EA FC 2025/male_players_EAFC25.csv')
        recommender = StrikerRecommender(df)
        recommended_players = recommender.recommend()
        print("Top 5 Striker Recommendations:")
        print(recommended_players)
    except Exception as e:
        logging.error(f"Main process error: {e}")

if __name__ == "__main__":
    main()
```
