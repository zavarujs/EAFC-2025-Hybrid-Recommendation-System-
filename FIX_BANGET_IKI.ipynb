```python
from sklearn import preprocessing

x = df.values
scaler = preprocessing.MinMaxScaler()
x_scaled = scaler.fit_transform(x)
X_norm = pd.DataFrame(x_scaled)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 2) # 2D PCA untuk visual plot
reduced = pd.DataFrame(pca.fit_transform(X_norm))
```

```python
from sklearn.cluster import KMeans

# tentukan jumlah cluster
kmeans = KMeans(n_clusters=5) # 5 cluster cuk wakeh men
# ngopo kok dadak 5 cluster yo mergo iki kan data set pemain
# pasti due 5 posisi pasti : forward, midfield, defender, winger, goalkepeer. ngeten cok

# sesuai dengan data masukan
kmeans = kmeans.fit(reduced)

# mendapatkan cluster labels
labels = kmeans.predict(reduced)

# centroid values, nggo nemukke centroid
centroid = kmeans.cluster_centers_

# cluster values
clusters = kmeans.labels_.tolist() # range 0-4 cluster (1-5 cluster)
```

```python
reduced['cluster'] = clusters
reduced['name'] = names
reduced.columns = ['x', 'y', 'cluster', 'name'] # x dan y itu untuk reduksi PCAne jadi di kurangi dadi 2 dimensi
# yaitu x dan y ben visual neng 2 rung demensi
reduced.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics.pairwise import cosine_similarity

class PlayerRecommendationSystem:
    def __init__(self, file_path):
        # Load dataset
        self.df = pd.read_csv(file_path)

        # Definisi bobot atribut untuk setiap posisi
        self.position_attributes = {
            'ST': ['OVR', 'Finishing', 'Shot Power', 'Positioning', 'Sprint Speed', 'Weak foot'],
            'CAM': ['OVR', 'Vision', 'Short Passing', 'Dribbling', 'Positioning', 'Weak foot'],
            'CDM': ['OVR', 'Interceptions', 'Standing Tackle', 'Stamina', 'PAS', 'Strength'],
            'CM': ['OVR', 'Short Passing', 'Long Passing', 'Vision', 'Stamina', 'Ball Control'],
            'LW': ['OVR', 'Dribbling', 'Sprint Speed', 'Crossing', 'Acceleration', 'Weak foot'],
            'RW': ['OVR', 'Dribbling', 'Sprint Speed', 'Crossing', 'Acceleration', 'Weak foot'],
            'CB': ['OVR', 'Strength', 'Heading Accuracy', 'Standing Tackle', 'Jumping', 'Aggression'],
            'LB': ['OVR', 'Stamina', 'DEF', 'Standing Tackle', 'Sprint Speed', 'Crossing'],
            'RB': ['OVR', 'Stamina', 'DEF', 'Standing Tackle', 'Sprint Speed', 'Crossing'],
            'GK': ['OVR', 'GK Diving', 'GK Handling', 'GK Positioning', 'GK Reflexes', 'Composure']
        }

    def recommend_players(self, input_data):
        # Filter pemain berdasarkan kriteria input
        filtered_df = self.df.copy()

        # Filter berdasarkan Overall Rating
        if 'OVR' in input_data:
            filtered_df = filtered_df[filtered_df['OVR'] >= input_data['OVR']]

        # Filter berdasarkan Posisi
        if 'Position' in input_data:
            position = input_data['Position']
            filtered_df = filtered_df[filtered_df['Position'] == position]

            # Pilih atribut berdasarkan posisi
            position_attrs = self.position_attributes.get(position,
                                                          ['OVR', 'PAC', 'SHO', 'PAS', 'DRI'])
        else:
            position_attrs = ['OVR', 'PAC', 'SHO', 'PAS', 'DRI']

        # Filter tambahan opsional
        optional_filters = {
            'Weak foot': lambda df, val: df[df['Weak foot'] >= val],
            'Skill moves': lambda df, val: df[df['Skill moves'] >= val],
            'Age': lambda df, val: df[df['Age'] >= val]
        }

        # Terapkan filter tambahan
        for key, filter_func in optional_filters.items():
            if key in input_data:
                filtered_df = filter_func(filtered_df, input_data[key])

        # Normalisasi data
        scaler = MinMaxScaler()

        # Siapkan fitur untuk content-based filtering
        content_features = filtered_df[position_attrs]
        content_features_scaled = pd.DataFrame(scaler.fit_transform(content_features),
                                               columns=position_attrs)

        # Hitung similarity
        user_profile = [input_data.get(attr, content_features[attr].mean()) for attr in position_attrs]
        content_similarities = cosine_similarity([user_profile], content_features_scaled)[0]

        # Collaborative filtering (simulasi sederhana)
        collab_similarities = np.random.rand(len(filtered_df))

        # Hybrid scoring
        hybrid_scores = (0.6 * content_similarities) + (0.4 * collab_similarities)

        # Tambahkan skor ke dataframe
        filtered_df['hybrid_score'] = hybrid_scores

        # Sorting dan top rekomendasi
        recommendations = filtered_df.sort_values('hybrid_score', ascending=False).head(10)

        # Kolom output
        output_columns = ['Name', 'Position', 'OVR'] + [attr for attr in position_attrs if attr != 'OVR']

        return recommendations[output_columns]

# Inisialisasi recommender
recommender = PlayerRecommendationSystem('/content/drive/MyDrive/SKRIPSI/EA FC 2025/male_players_EAFC25.csv')

# Fungsi untuk input manual
def get_recommendation():
    print("Sistem Rekomendasi Pemain Sepak Bola")
    input_data = {}

    # Input opsional
    inputs = [
        ('OVR', int, "Minimal Overall Rating"),
        ('Position', str, "Posisi Pemain (ST/CAM/CDM/CM/LW/RW/CB/LB/RB/GK)"),
        ('Weak foot', int, "Minimal Weak Foot (1-5)"),
        ('Skill moves', int, "Minimal Skill Moves (1-5)"),
        ('Age', int, "Minimal Umur")
    ]

    for prompt, data_type, description in inputs:
        while True:
            try:
                value = input(f"Masukkan {description} (kosongkan jika tidak perlu): ")
                if value:
                    input_data[prompt] = data_type(value)
                break
            except ValueError:
                print("Input tidak valid. Coba lagi.")

    # Dapatkan rekomendasi
    recommendations = recommender.recommend_players(input_data)

    # Tampilkan rekomendasi
    print("\n--- Rekomendasi Pemain ---")
    print(recommendations)

# Jalankan rekomendasi
get_recommendation()
```
